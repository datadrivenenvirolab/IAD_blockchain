```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("V2_COMPLETE_DATA_COUNTRY_4TOPICS.csv", index_col=0)
```

```python
df.head(20)
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x = df[['Topic1', 'Topic2', 'Topic3', 'Topic4']]
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
```

```python
sum(pca.explained_variance_ratio_)
```

```python
final = pd.concat([df,principalDf], axis=1)
```

```python
final
```

```python
final.to_csv("V2_PCA_RAW_DATA.csv")
```

## T-SNE

```python
from sklearn.manifold import TSNE
tsne_model = TSNE(n_components=2, verbose=1, random_state=0, angle=.99, init='pca', perplexity=30)
tsne_lda = tsne_model.fit_transform(x)

tsne_data = pd.DataFrame(data = tsne_lda,
                           columns = ['X', 'Y'])
```

```python
final_tsne = pd.concat([df,tsne_data], axis=1)
```

```python
final_tsne
```

```python
F =final_tsne.copy()
F = F.rename(columns = {'X': 'tsne_pc1', 'Y':'tsne_pc2'})
F
```

```python
F.to_csv("V2_TSNE_RAW_DATA_perp30.csv")
```
