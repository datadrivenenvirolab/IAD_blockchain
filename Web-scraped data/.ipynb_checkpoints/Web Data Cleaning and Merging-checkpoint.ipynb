## Merging all data frames into one

```python
df1 = pd.read_csv("scraped_websites_0_100.csv", index_col=0)
df2 = pd.read_csv("scraped_websites_100_200.csv", index_col=0)
df3 = pd.read_csv("scraped_websites_200_300.csv", index_col = 0)

df = pd.concat([df1, df2, df3], ignore_index=True)

i = 299
while(i<=1049):
    current_df = pd.read_csv(f"scraped_websites_{i}_{i+50}.csv", index_col=0)
    df = pd.concat([df, current_df], ignore_index=True)
    i = i + 50

final_df = pd.read_csv("scraped_websites_last_stretch.csv", index_col=0)
df = pd.concat([df, final_df], ignore_index=True)

df = pd.concat([df, unscraped], ignore_index=True)
df
```

## Unicode Normalization (for \xa characters)

```python
import unicodedata

def unicode_normalize (given_string):
    return unicodedata.normalize("NFKD", given_string)

df.Text = df.Text.map(unicode_normalize)
```

# Pattern 1 to Clean: small letter + Big letter

```python
txt = "2020Integrated policy is the 2019Best"

print (re.sub(r'(?<=[0-9])(?=[A-Z])', ' ', txt))
```

```python
df.Text = df.Text.map(lambda txt: re.sub(r'(?<=[0-9])(?=[A-Z])', ' ', txt))
```

# Pattern 3i to Clean:  ] + Big letter

```python
df.Text = df.Text.map(lambda txt: re.sub(r'(?<=\])(?=[A-Z])', ' ', txt))
```

# Pattern 3ii to Clean: small letter + [

# Pattern 4 to Clean: ) + Big letter

```python
txt = "(IIED)Founded1973; 48 years ago (1973)Founder Barbara Ward Type"
print (re.sub(r'(?<=\))(?=[A-Z])', ' ', txt))
```

```python
df.Text = df.Text.map(lambda txt: re.sub(r'(?<=\))(?=[A-Z])', ' ', txt))
```

# Pattern 5 to Clean: small letter + number

```python
txt = "Founded1973 made1890 it is there"
print (re.sub(r'(?<=[a-z])(?=[0-9])', ' ', txt))
```

```python
df.Text = df.Text.map(lambda txt: re.sub(r'(?<=[a-z])(?=[0-9])', ' ', txt))
```

# Pattern 6 to Clean: small letter + (

# Remove any links

```python
import re
# from df.Text[3233]
text = "International Institute for Environment and Development. www.iied.org. ^ IIED finances [1] ^ https://www.af-info.or.jp/en/blueplanet/doc/list/1992essay-iied.pdf ^ a b c Batterbury, SPJ. 2004. The International Institute for Environment and Development: notes on a small office. Global Environmental Change 14: 367–371. ^ Ed Nigel Cross"
print(re.sub(r'http\S+', '', text))
```

```python
txt = "this is [1] a cat [edit] here"
print(re.sub(r"(\[\S*\])", "", txt))
```

```python
import pycld2 as cld2

isReliable, textBytesFound, details = cld2.detect(
    "а неправильный формат идентификатора дн назад"
)
print(details[0])
```

```python
list_of_languages = []
for text in df.Text:
    try:
        isReliable, textBytesFound, details = cld2.detect(text)
        language = str(details[0][0])
        list_of_languages.append(language)
    except:
        list_of_languages.append("Undetected")

df["original_language"] = list_of_languages
df
```
