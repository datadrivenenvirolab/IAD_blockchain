```python
def transform_constituency(const):
    d = {"Constituency - Business and Industry":"BINGO",
        "Constituency - Business and Industry2":"BINGO",
        "Constituency - ENGO":"ENGO",
        "Constituency - Environmental":"ENGO",
        "Constituency - Environmental CAN":"ENGO",
        "Constituency - Environmental CJN":"ENGO",
        "Constituency - Farmers":"Farmers",
        "Constituency - Indigenous peoples":"IPO",
        "Constituency - Local government and Municipal authorities":"LGMA",
        "Constituency - Research and independent":"RINGO",
        "Constituency - Trade Unions":"TUNGO",
        "Constituency - Women and Gender":"Women and Gender",
        "Constituency - Youth":"Youth",
        "not yet affiliated/no data":"not yet affiliated/no data"}
    const = list(const)
    new = []
    for c in const:
        try:
            new.append(d[c])
        except:
            new.append("not yet affiliated/no data")  
    return new
```

```python
def generate_constituency_stats(NGO):
    NGO_constituency_stats = pd.DataFrame(NGO.const.value_counts())
    NGO_constituency_stats.reset_index(level=0, inplace=True)
    NGO_constituency_stats.columns = ["Constituency", "Count"]
    NGO_constituency_stats.sort_values(by = "Constituency", inplace=True)
    return NGO_constituency_stats

cstats_2015 = generate_constituency_stats(ngo_2015)
cstats_2015.to_csv("COP2015_constituency_stats_new.csv")

cstats_2017 = generate_constituency_stats(ngo_2017)
cstats_2017.to_csv("COP2017_constituency_stats_new.csv")

cstats_2019 = generate_constituency_stats(ngo_2019)
cstats_2019.to_csv("COP2019_constituency_stats_new.csv")

cstats_2021 = generate_constituency_stats(ngo_2021)
cstats_2021.to_csv("COP2021_constituency_stats_new.csv")
```

```python
def refactor_dfs_perc(dfs, cols):
    init = pd.DataFrame(columns=cols)
    for i in range(len(dfs)):
        df = dfs[i]
        a = df[["Constituency", "Count"]].copy().T
        a.reset_index(drop = True, inplace=True)
        a.columns = a.iloc[0,:]
        a = a.drop(labels=0, axis=0)
        a.reset_index(drop = True, inplace=True)
        a["Conference"] = [f"COP{2015+2*i}"]
        init = init.append(a, ignore_index=True)
    
    return init
```

```python
cols = list(cstats_2015.Constituency)
dfs = [cstats_2015, cstats_2017, cstats_2019, cstats_2021]
aggregated = refactor_dfs_perc(dfs, cols)
```

# Topic Heatmap

```python
complete_data = pd.read_csv("V2_COMPLETE_DATA_COUNTRY_4TOPICS.csv", index_col=0)
complete_data = complete_data[complete_data.entity_type == "Non-governmental organizations"]
complete_data = complete_data[['organization', 'most_prevalent_topic', 'continent', 'income_group']]
complete_data.reset_index(drop = True, inplace = True)
complete_data
```

```python
names_const = ngo_2019.copy()[['name', 'constituency']]
names_const = names_const[names_const.name!='Institute de Recherche pour le Developpement']
names_const.reset_index(drop = True, inplace = True)
```

```python
names_const
```

```python
# set(names_const.name) - set(complete_data.organization)
```

```python
# pd.merge(restaurant_ids_dataframe, restaurant_review_frame, on='business_id', how='outer')
names_const.rename({'name': 'organization'}, inplace=True, axis = 1)
```

```python
names_const
```

```python
merged = pd.merge(complete_data, names_const, on='organization', how='outer')
merged
```

```python
df = merged.groupby(by = 'constituency').agg('count')
df
```

```python
dummies = pd.get_dummies(data = merged.most_prevalent_topic, columns = ['topic1_prevalent','topic2_prevalent', 'topic3_prevalent', 'topic4_prevalent'])
dummies
```

```python
new_merged = pd.concat([merged, dummies], axis = 1)
df = new_merged.groupby(by = ['constituency'])[1, 2, 3, 4].agg('sum')
df.rename({1:'topic 1', 2:'topic 2', 3:'topic 3', 4:'topic 4'}, inplace=True, axis = 1)
df
```

```python
fig, ax = plt.subplots(figsize=(12,12)) 


sns.heatmap(df, 
            cmap="YlGnBu",
            linewidths=.5,
            annot = True,
            fmt = 'g',
            vmin = 0,
            vmax = 160,
            ax = ax)
ax.xaxis.tick_top() # x axis on top
ax.xaxis.set_label_position('top')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('NGO_constituency_vs_topic_heatmap.png', facecolor = "white", dpi = 350)
```

# Re-doing side-by-side barplots for 18 topic model

```python
from gensim import corpora, models, similarities
path = "/Users/stefan/Desktop/IAD_blockchain/Web-scraped data/V2 Final Models/18 topics/V2_18topics_model"
lda_model_18t = models.LdaModel.load(path)
```

```python
def unzip_list_of_tuples(L):
    return [x[0] for x in L], [x[1] for x in L]

def transform_topic_word_list_to_df(topic_word_list):
    result = []
    for i in range(len(topic_word_list)):
        current = int(topic_word_list[i][0])
        current = current + 1
        words, scores = unzip_list_of_tuples(topic_word_list[i][1])
        topics = [current] * len(words)
        word_scores_topic = list(zip(words,scores,topics))
        result.extend(word_scores_topic)
    return pd.DataFrame(result, columns = ["word", "score", "topic"])
```

```python
eighteen_topic = list(lda_model_18t.show_topics(18, num_words=10, formatted=False))
eighteen_topic_df = transform_topic_word_list_to_df(eighteen_topic)
eighteen_topic_df
```

```python
import matplotlib.pyplot as plt

def generate_side_by_side_barplots(df):
    undesirable_topics = [1, 2, 3, 5, 6, 12, 17]
    title_labels = {4:"Topic 4: Subnational Governments",
                    7:"Topic 7: Agriculture and Forests",
                    8:"Topic 8: Energy",
                    9:"Topic 9: University",
                    10:"Topic 10: Development, Community and Underrepresented Groups",
                    11:"Topic 11: Global South",
                    13:"Topic 13: IGO",
                    14:"Topic 14: ENGO",
                    15:"Topic 15: Sustainable Development",
                    16:"Topic 16: International",
                    18:"Topic 18: Private Sector"}
    
    plt.style.use('dark_background')
    
    plt.figure(figsize=(15,15))
    j = 1
    for i in range(1,19):
        if i not in undesirable_topics:
            t = df[df.topic==i].copy()
            plt.subplot(4,3,j)
            j = j + 1
            plt.tight_layout()
            plt.gca().invert_yaxis()
            plt.xlabel("Word Weight")
            plt.title(title_labels[i])
            plt.barh(list(t.word), list(t.score))
    plt.savefig("V2_18topic_model_word_barplots_cut.png", facecolor = "black", dpi = 300)
    plt.show()
```

```python
generate_side_by_side_barplots(eighteen_topic_df)
```
