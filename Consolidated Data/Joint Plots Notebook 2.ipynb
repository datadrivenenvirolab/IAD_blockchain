```python
def refactor_dfs_perc(dfs, cols):
    init = pd.DataFrame(columns=cols)
    for i in range(len(dfs)):
        df = dfs[i]
        a = df[["Constituency", "Count"]].copy().T
        a.reset_index(drop = True, inplace=True)
        a.columns = a.iloc[0,:]
        a = a.drop(labels=0, axis=0)
        a.reset_index(drop = True, inplace=True)
        a["Conference"] = [f"COP{2015+2*i}"]
        init = init.append(a, ignore_index=True)
    
    return init
```

```python
cols = list(cstats_2015.Constituency)
dfs = [cstats_2015, cstats_2017, cstats_2019, cstats_2021]
aggregated = refactor_dfs_perc(dfs, cols)
```

# Topic Heatmap

```python
complete_data = pd.read_csv("V2_COMPLETE_DATA_COUNTRY_4TOPICS.csv", index_col=0)
complete_data = complete_data[complete_data.entity_type == "Non-governmental organizations"]
complete_data = complete_data[['organization', 'most_prevalent_topic', 'continent', 'income_group']]
complete_data.reset_index(drop = True, inplace = True)
complete_data
```

```python
names_const = ngo_2019.copy()[['name', 'constituency']]
names_const = names_const[names_const.name!='Institute de Recherche pour le Developpement']
names_const.reset_index(drop = True, inplace = True)
```

```python
names_const
```

```python
# set(names_const.name) - set(complete_data.organization)
```

```python
# pd.merge(restaurant_ids_dataframe, restaurant_review_frame, on='business_id', how='outer')
names_const.rename({'name': 'organization'}, inplace=True, axis = 1)
```

```python
names_const
```

```python
merged = pd.merge(complete_data, names_const, on='organization', how='outer')
merged
```

```python
df = merged.groupby(by = 'constituency').agg('count')
df
```

```python
dummies = pd.get_dummies(data = merged.most_prevalent_topic, columns = ['topic1_prevalent','topic2_prevalent', 'topic3_prevalent', 'topic4_prevalent'])
dummies
```

```python
new_merged = pd.concat([merged, dummies], axis = 1)
df = new_merged.groupby(by = ['constituency'])[1, 2, 3, 4].agg('sum')
df.rename({1:'topic 1', 2:'topic 2', 3:'topic 3', 4:'topic 4'}, inplace=True, axis = 1)
df
```

```python
fig, ax = plt.subplots(figsize=(12,12)) 


sns.heatmap(df, 
            cmap="YlGnBu",
            linewidths=.5,
            annot = True,
            fmt = 'g',
            vmin = 0,
            vmax = 160,
            ax = ax)
ax.xaxis.tick_top() # x axis on top
ax.xaxis.set_label_position('top')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('NGO_constituency_vs_topic_heatmap.png', facecolor = "white", dpi = 350)
```
